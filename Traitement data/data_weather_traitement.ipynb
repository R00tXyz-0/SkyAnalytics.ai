```python
# Importer la bibiothèque
import pandas as pd
```

```python
# Loading data
df_c = pd.read_csv("casa_weather.csv")
df_r = pd.read_csv("rabat_weather.csv")
```

```python
df_c.info()
```

```python
df_c["time"] = pd.to_datetime(df_c["time"])

df_r["time"] = pd.to_datetime(df_r["time"])
```

```python
df_c.info()
```

```python
df_c["year"] = df_c["time"].dt.year
df_c["month"] = df_c["time"].dt.month
df_c["day"] = df_c["time"].dt.day
```

```python
df_r["year"] = df_r["time"].dt.year
df_r["month"] = df_r["time"].dt.month
df_r["day"] = df_r["time"].dt.day
```

```python
df_c.columns
```

```python
df_c["temperature"] = (
    df_c.groupby(df_c["day"])["temperature"].transform("mean")
)
df_c["humidity"] = (
    df_c.groupby(df_c["day"])["humidity"].transform("mean")
)
df_c["precipitation"] = (
    df_c.groupby(df_c["day"])["precipitation"].transform("mean")
)
df_c["cloudcover"] = (
    df_c.groupby(df_c["day"])["cloudcover"].transform("mean")
)
df_c["windspeed"] = (
    df_c.groupby(df_c["day"])["windspeed"].transform("mean")
)
```

```python
df_r["temperature"] = (
    df_r.groupby(df_r["day"])["temperature"].transform("mean")
)
df_r["humidity"] = (
    df_r.groupby(df_r["day"])["humidity"].transform("mean")
)
df_r["precipitation"] = (
    df_r.groupby(df_r["day"])["precipitation"].transform("mean")
)
df_c["cloudcover"] = (
    df_r.groupby(df_r["day"])["cloudcover"].transform("mean")
)
df_r["windspeed"] = (
    df_r.groupby(df_r["day"])["windspeed"].transform("mean")
)
```

```python

df_c.head()
```

```python
df_c = df_c.drop_duplicates(subset=["year", "month", "day"])
df_r = df_r.drop_duplicates(subset=["year", "month", "day"])
```

```python
df_c = df_c.drop("time", axis=1)
df_r = df_r.drop("time", axis=1)
```

```python
df_c.to_csv("data_traiter_casa.csv", index=False)
df_r.to_csv("data_traiter_rabat.csv", index=False)
```
